### What factors (e.g., study time, parental education) most influence student grades in math and Portuguese courses?

#### First, let's check how the other columns correlates to grades in each course
#### In this instance, we'll use the merged dataset
#### In order to use this I'll have to filter out non-numeric datatypes

```python
dfNumeric = dfMerge.select_dtypes(include=['int64', 'float64'])
```

#### Note that g3 is the final grade, so there's no need to look at g2 and g1
#### Also, the correlation will be represented by a number from 1 to -1. Numbers closer to 0 mean no correlation, closer to 1 or -1 indicate a strong correlation

```python
correlations = dfNumeric.corr()['G3'].sort_values(ascending=False)
print(correlations)
```

### Visualization of correlation via heatmap (courtsey of chatGPT)
#### dropping g3, g2, and g1

```python
column_rename = {
    'Medu': "Mother's Education Level",
    'studytime': 'Weekly Study Time',
    'Fedu': "Father's Education Level",
    'famrel': 'Family Relationship Quality',
    'absences': 'Number of Absences',
    'freetime': 'Free Time After School',
    'health': 'Current Health Status',
    'goout': 'Going Out Frequency',
    'traveltime': 'Travel Time to School',
    'Walc': 'Weekend Alcohol Consumption',
    'age': 'Age of Student',
    'Dalc': 'Weekday Alcohol Consumption',
    'failures': 'Number of Past Failures'
}

correlation_df = correlations.drop(labels=['G1', 'G2', 'G3']).reset_index()

correlation_df['index'] = correlation_df['index'].replace(column_rename)
correlation_df.columns = ['Feature', 'Correlation']

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_df.set_index('Feature'), annot=True, fmt=".2f", cmap='coolwarm', cbar=True)
plt.title("Parent's Education Level, Time Spent Studying, and Number of Past Failures are the Greatest Indicators of Success")
plt.show()
```

### I'm curious what combination of two factors lead to a hight final grade. To help answer this question I used chatGPT

```python
from itertools import combinations

numerical_cols = dfMerge.select_dtypes(include='number').drop(columns=['G1', 'G2', 'G3'])

feature_combinations = list(combinations(numerical_cols.columns, 2))

interaction_correlations = {}
for feature1, feature2 in feature_combinations:
    interaction_term = dfMerge[feature1] * dfMerge[feature2]
    correlation = interaction_term.corr(dfMerge['G3'])
    interaction_correlations[f"{feature1} * {feature2}"] = correlation

sorted_interactions = sorted(interaction_correlations.items(), key=lambda x: abs(x[1]), reverse=True)

for interaction, corr in sorted_interactions[:5]:
    print(f"Interaction: {interaction}, Correlation with G3: {corr:.2f}")
```

```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import numpy as np

top_interactions = sorted_interactions[:5]

renamed_interactions = [
    f"{column_rename[term.split(' * ')[0]]} * {column_rename[term.split(' * ')[1]]}" 
    for term, _ in top_interactions
]

correlations = [corr for _, corr in top_interactions]

num_bars = len(correlations)
base_color = '#1f77b4'  
colors = [
    mcolors.to_rgba(base_color, alpha) 
    for alpha in np.linspace(1, 0.5, num_bars)
]

plt.figure(figsize=(10, 6))
plt.barh(renamed_interactions, correlations, color=colors)

plt.title('Top 5 Combination of Two Factors Correlated with Final Grade', fontsize=14)
plt.xlabel('Correlation Coefficient', fontsize=12)
plt.ylabel('Combination of Two Factors', fontsize=12)

plt.gca().invert_yaxis()

for index, corr in enumerate(correlations):
    plt.text(corr, index, f"{corr:.2f}", va='center', ha='left' if corr > 0 else 'right', color='black')

plt.tight_layout()
plt.show()
```

#### It seems like past failures are the greatest indicator of student success
#### This means that schools should provide students who have failed in the past with additional assistence in order to boost overall grades

### Are there noticeable differences in performance between male and female students for Porteguese and Math?

#### Group data by gender and calculate mean grades for both datasets

```python
genderMath = dfMat.groupby('sex')['G3'].mean()
genderPor = dfPor.groupby('sex')['G3'].mean()
```

#### Put the results in a dataframe

```python
genderComp = pd.DataFrame({
    'Male': [genderMath['M'], genderPor['M']],
    'Female': [genderMath['F'], genderPor['F']]
}, index=['Math', 'Portuguese'])
```

#### Graph the gender comparision (chatGPT)

```python
genderComp.plot(kind='bar', figsize=(10, 6), color=['blue', 'pink'])
plt.title('Average Final Grades by Subject for Each Gender')
plt.ylabel('Average Grade')
plt.xlabel('Subject')
plt.xticks(rotation=0)
plt.legend(title='Gender')
plt.show()
```

#### Males tend to be more successful in math, while females tend to be more successful in Portuguese
